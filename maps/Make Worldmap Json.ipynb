```python
import json
import csv


csvfile = open('data/worldmap.csv', 'r+')
jsonfile = open('data/worldmap.json', 'w+')

fieldnames = ("homelon","homelat","homecontinent","n")

reader = csv.DictReader( csvfile, fieldnames)
jsonfile.write('[')
for row in reader:
    json.dump(row, jsonfile)
    jsonfile.write(',\n')
jsonfile.write(']')

csvfile.close()
jsonfile.close()

    
```
